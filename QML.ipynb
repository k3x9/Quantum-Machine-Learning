```python
def batchLoader(myDir, batchsize = 100, clsses = 2):
    X = []
    Y = []
    X_t = []
    Y_t = []
    import os
    lst = []
    lst1 = []
    cnt = 0
    for i in os.listdir(myDir):
        cnt += 1
        if (cnt > clsses):
            break
        currDir = myDir +  i
        
        sz = len(os.listdir(currDir))
        for j in range(batchsize//clsses):
            a = np.random.randint(sz)
            while a in lst:
                a = np.random.randint(sz)
            lst.append(currDir + '/' + str(a) + '.png')
        for j in range(10):
            a = np.random.randint(sz)
            while a in lst:
                a = np.random.randint(sz)
            lst1.append(currDir + '/' + str(a) + '.png')
    
    for file in lst:
        img_file = Image.open(file)
        img_file = img_file.resize((3, 3), resample=Image.BILINEAR)
        width, height = img_file.size
        format = img_file.format
        mode = img_file.mode
        img_grey = img_file.convert('L')
        value = np.asarray(img_grey.getdata()).reshape((width, height))
        value = value.flatten()


        X.append(value)

        if file.split('/')[-2] == 'circle':
            Y.append(0)
        elif file.split('/')[-2] == 'square':
            Y.append(1)
        elif file.split('/')[-2] == 'triangle':
            Y.append(2)
        else:
            Y.append(3)
    for file in lst1:
        img_file = Image.open(file)
        img_file = img_file.resize((3, 3), resample=Image.BILINEAR)
        width, height = img_file.size
        format = img_file.format
        mode = img_file.mode
        img_grey = img_file.convert('L')
        value = np.asarray(img_grey.getdata()).reshape((width, height))
        value = value.flatten()


        X_t.append(value)

        if file.split('/')[-2] == 'circle':
            Y_t.append(0)
        elif file.split('/')[-2] == 'square':
            Y_t.append(1)
        elif file.split('/')[-2] == 'triangle':
            Y_t.append(2)
        else:
            Y_t.append(3)
    return np.array(X), np.array(Y), np.array(X_t), np.array(Y_t)


```

Quantum Covolution layers

```python
def conv_circuit(params):
    target = QuantumCircuit(2)
    target.rz(-np.pi / 2, 1)
    target.cx(1, 0)
    target.rz(params[0], 0)
    target.ry(params[1], 1)
    target.cx(0, 1)
    target.ry(params[2], 1)
    target.cx(1, 0)
    target.rz(np.pi / 2, 0)
    return target
```

```python
params = ParameterVector("θ", length=3)
circuit = conv_circuit(params)
circuit.draw("mpl")
```

```python
def conv_layer(num_qubits, param_prefix):
    qc = QuantumCircuit(num_qubits, name="Convolutional Layer")
    qubits = list(range(num_qubits))
    param_index = 0
    params = ParameterVector(param_prefix, length=num_qubits * 3)
    for q1, q2 in zip(qubits[0::2], qubits[1::2]):
        qc = qc.compose(conv_circuit(params[param_index : (param_index + 3)]), [q1, q2])
        qc.barrier()
        param_index += 3
    for q1, q2 in zip(qubits[1::2], qubits[2::2] + [0]):
        qc = qc.compose(conv_circuit(params[param_index : (param_index + 3)]), [q1, q2])
        qc.barrier()
        param_index += 3

    qc_inst = qc.to_instruction()

    qc = QuantumCircuit(num_qubits)
    qc.append(qc_inst, qubits)
    return qc
```

```python
circuit = conv_layer(4, "θ")
circuit.decompose().draw("mpl")
```

```python
def pool_circuit(params):
    target = QuantumCircuit(2)
    target.rz(-np.pi / 2, 1)
    target.cx(1, 0)
    target.rz(params[0], 0)
    target.ry(params[1], 1)
    target.cx(0, 1)
    target.ry(params[2], 1)

    return target
```

```python
params = ParameterVector("θ", length=3)
circuit = pool_circuit(params)
circuit.draw("mpl")
```

```python
def pool_layer(sources, sinks, param_prefix):
    num_qubits = len(sources) + len(sinks)
    qc = QuantumCircuit(num_qubits, name="Pooling Layer")
    param_index = 0
    params = ParameterVector(param_prefix, length=num_qubits // 2 * 3)
    for source, sink in zip(sources, sinks):
        qc = qc.compose(pool_circuit(params[param_index : (param_index + 3)]), [source, sink])
        qc.barrier()
        param_index += 3

    qc_inst = qc.to_instruction()

    qc = QuantumCircuit(num_qubits)
    qc.append(qc_inst, range(num_qubits))
    return qc
```

```python
sources = [0, 1]
sinks = [2, 3]
circuit = pool_layer(sources, sinks, "θ")
circuit.decompose().draw("mpl")
```

```python
feature_map = ZFeatureMap(9)
feature_map.decompose().draw("mpl")
```

```python
convo = QuantumCircuit(9, name="Convo")


convo.compose(conv_layer(9, "с1"), list(range(9)), inplace=True)
convo.compose(pool_layer([0, 1, 2, 3], [4, 5, 6, 7, 8], "p1"), list(range(9)), inplace=True)
convo.compose(conv_layer(4, "c2"), list(range(5, 9)), inplace=True)
convo.compose(pool_layer([0, 1], [2, 3], "p2"), list(range(5, 9)), inplace=True)


```

```python
circuit = QuantumCircuit(9)
circuit.compose(feature_map, range(9), inplace=True)
circuit.compose(convo, range(9), inplace=True)
```

```python
circuit.decompose().draw(output="mpl")
```

```python
objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

```python
sampler_qnn = SamplerQNN(
    circuit=circuit,
    input_params=feature_map.parameters,
    weight_params=convo.parameters,
    interpret=None,
    output_shape=output_shape,
)
```

```python
classifier = NeuralNetworkClassifier(
    neural_network = sampler_qnn,
    optimizer=COBYLA(maxiter=300),
    callback=callback_graph
)
```
